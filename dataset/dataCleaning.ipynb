```python
import pandas as pd

df = pd.read_csv("Global_Landslide_Catalog_Export_rows (1).csv")

# Clean the date column
df['event_date'] = pd.to_datetime(df['event_date'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')

# Drop rows with missing coordinates or dates
df = df.dropna(subset=['latitude', 'longitude', 'event_date'])

# Filter to Himachal Pradesh bounding box
hp = df[
    (df['latitude']  >= 30.2) & (df['latitude']  <= 33.3) &
    (df['longitude'] >= 75.5) & (df['longitude'] <= 79.0) &
    (df['event_date'] >= '2015-01-01') &
    (df['event_date'] <= '2022-12-31')
].copy()

hp['event_date'] = hp['event_date'].dt.strftime('%Y-%m-%d')

print(f"HP landslide events found: {len(hp)}")
print(hp[['latitude', 'longitude', 'event_date', 'location_description']].to_string())
```

```python
import pandas as pd

df = pd.read_csv("Global_Landslide_Catalog_Export_rows (1).csv")

df['event_date'] = pd.to_datetime(df['event_date'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')
df = df.dropna(subset=['latitude', 'longitude', 'event_date'])

hp = df[
    (df['latitude']  >= 30.2) & (df['latitude']  <= 33.3) &
    (df['longitude'] >= 75.5) & (df['longitude'] <= 79.0) &
    (df['event_date'] >= '2015-01-01') &
    (df['event_date'] <= '2022-12-31')
].copy()

hp['event_date'] = hp['event_date'].dt.strftime('%Y-%m-%d')

# Save only the 3 columns GEE needs
hp[['latitude', 'longitude', 'event_date']].to_csv("hp_landslides_upload.csv", index=False)
print(f"Saved {len(hp)} rows to hp_landslides_upload.csv")
```
